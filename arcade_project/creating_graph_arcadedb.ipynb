````python
def extract_graph_schema_from_text(text: str) -> Dict:
    """
    Extract entities and relationships with their types and properties.
    """
    prompt = f"""
Given the following text:
\"\"\"{text}\"\"\"

Extract:
1. A list of entities, with each entity having:
   - name
   - type
   - properties (as key-value pairs)

2. A list of relationships between entities, with:
   - source entity name
   - target entity name
   - type of relation
   - properties (if any)

Only Return the result as a valid JSON object, and should not contain any other text or ''', 
should not contain any headings or language or format-types( like json )and details , 
i need a clean json object that should start with a curly brackets and end with curly braces 

example output format:
{{
  "entities": [
    {{
      "name": "Uday",
      "type": "Person",
      "properties": {{
        "age": 28,
        "city": "Hyderabad"
      }}
    }},
    ...
  ],
  "relations": [
    {{
      "source": "Uday",
      "target": "TCS",
      "type": "WORKS_IN",
      "properties": {{
        "since": 2020
      }}
    }},
    ...
  ]
}}
"""
    output = llm._call(prompt)
    try:
        # Extract JSON code block between triple backticks (``` or ```json)
        code_block_match = re.search(r"```(?:json)?(.*?)```", output, re.DOTALL)
        if code_block_match:
            cleaned = code_block_match.group(1).strip()
        else:
            # If no code block found, try to parse raw string directly
            cleaned = output.strip()

        # Load the cleaned string as JSON
        return json.loads(cleaned)
    
    except Exception as e:
        print(" Failed to parse JSON from LLM:", e)
        print("Raw Output:", output)
        return {"entities": [], "relations": []}
````

````python
import json

class GraphTypeCache:
    def __init__(self, llm, similarity_threshold=0.8):
        self.vertex_types = set()
        self.edge_types = set()
        self.llm = llm
        self.similarity_threshold = similarity_threshold
        logging.info("GraphTypeCache initialized with similarity_threshold=%.2f", similarity_threshold)

    def load_from_db(self):
        global run_cypher
        logging.info("Loading vertex and edge types from database...")

        try:
            label_query = "MATCH (n) RETURN DISTINCT labels(n) AS labels"
            label_result = run_cypher(label_query)
            self.vertex_types = {
                label for row in label_result['result'] for label in row['labels']
            }
            logging.info("Loaded vertex types: %s", self.vertex_types)

            rel_query = "MATCH ()-[r]->() RETURN DISTINCT type(r) AS relationship_type"
            rel_result = run_cypher(rel_query)
            self.edge_types = {
                row['relationship_type'] for row in rel_result['result']
            }
            logging.info("Loaded edge types: %s", self.edge_types)

        except Exception as e:
            logging.error("Failed to load types from database: %s", e)

    def _llm_match(self, new_type, existing_types, type_kind):
        if not existing_types:
            return None

        prompt = f"""
You are a smart semantic matcher for a graph database.

Check if the NEW {type_kind} TYPE is equivalent to any of the EXISTING {type_kind.upper()} TYPES.

NEW TYPE: '{new_type}'
EXISTING TYPES: {list(existing_types)}

Your task:
- If the new type is just a variation, abbreviation, misspelling, or synonym of an existing type, return the CLOSEST MATCH.
- If it is a new type entirely, return 'NONE'.

Only return:
- An EXACT existing type from the list above, or
- 'NONE'

Answer ONLY with the type name or 'NONE'.
"""
        logging.info("[LLM PROMPT]:\n%s", prompt)
        response = self.llm.invoke(prompt).strip().strip('"\'')
        logging.info("[LLM RESPONSE]: %s", response)

    # Normalize and check if it's a known type
        if response in existing_types:
            return response
        else:
            logging.warning("LLM response not in existing types: '%s'", response)
            return None



    def find_closest_vertex_type(self, label: str) -> str:
        logging.info("Finding closest vertex type for '%s'", label)
        match = self._llm_match(label, self.vertex_types, "vertex")
        if match:
            logging.info("Vertex type '%s' matched to existing '%s'", label, match)
        else:
            logging.info("Vertex type '%s' is treated as new", label)
        return match if match else label

    def find_closest_edge_type(self, rel: str) -> str:
        logging.info("Finding closest edge type for '%s'", rel)
        match = self._llm_match(rel, self.edge_types, "edge")
        if match:
            logging.info("Edge type '%s' matched to existing '%s'", rel, match)
        else:
            logging.info("Edge type '%s' is treated as new", rel)
        return match if match else rel

    def add_vertex_type(self, label: str):
        logging.info("Adding new vertex type to cache: '%s'", label)
        self.vertex_types.add(label)

    def add_edge_type(self, rel: str):
        logging.info("Adding new edge type to cache: '%s'", rel)
        self.edge_types.add(rel)
````

````python
def normalize(text: str) -> str:
    return text.lower().strip()

def map_graph_schema_with_cache(graph):
    updated_graph = {'entities': [], 'relations': []}

    # Process entities (vertex types)
    for entity in graph.get('entities', []):
        original_type = normalize(entity['type'])
        mapped_type = cache.find_closest_vertex_type(original_type)

        if mapped_type == original_type:
            cache.add_vertex_type(mapped_type)

        entity['type'] = mapped_type
        updated_graph['entities'].append(entity)

    # Process relations (edge types)
    for rel in graph.get('relations', []):
        original_type = normalize(rel['type'])
        mapped_type = cache.find_closest_edge_type(original_type)

        if mapped_type == original_type:
            cache.add_edge_type(mapped_type)

        rel['type'] = mapped_type
        updated_graph['relations'].append(rel)

    return updated_graph
````
